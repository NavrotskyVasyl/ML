```python
from scipy import stats
import numpy as np
import math as m
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k = [10,15,20,25,30]
n = 200
bins = 50
```

```python
for example in k:
    mu = example
    std = m.sqrt(example)
    x = np.random.normal(mu, std, n)
    y = np.random.chisquare(example, n)
    stat = stats.ks_2samp(x, y)

    plt.figure(figsize=(12,8))
    plt.title("k = %d, p-value = %f" % (example, stat.pvalue))
    plt.hist(x, bins, alpha=0.5, label='normal');
    plt.hist(y, bins, alpha=0.5, label='chi square');
    plt.legend(loc='upper left');
    plt.show()
```
